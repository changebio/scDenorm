```python
def download_cbd(name,file='exprMatrix.tsv.gz'):
    url = f"https://cells.ucsc.edu/{name}/{file}"
    filename = f"data/cellbrowserdatasets/{name}/{file}"
    folder_name = f"data/cellbrowserdatasets/{name}"

    try:
        response = urllib.request.urlopen(url)
        if response.status == 200:
            if not os.path.exists(folder_name):
                os.makedirs(folder_name)
            urllib.request.urlretrieve(url, filename)
        else:
            print(f"{name} Failed to download the file.")
    except:
        print(f'{name}. No connetion')
```

```python
def download_cbd1(name,file='exprMatrix.tsv.gz'):
    url = f"https://cells.ucsc.edu/{name}/{file}"
    filename = f"data/cellbrowserdatasets/{name}/{file}"
    folder_name = f"data/cellbrowserdatasets/{name}"

    try:
        response = urllib.request.urlopen(url)
    except:
        try:
            response = requests.get(url,timeout=5)
        except:
            print('No connetion')
    # Check if the request was successful
    if response.status == 200:
        if not os.path.exists(folder_name):
            os.makedirs(folder_name)
        urllib.request.urlretrieve(url, filename)
    elif response.status_code == 200:
        if not os.path.exists(folder_name):
            os.makedirs(folder_name)
        # Open a file in binary mode and write the response content
        with open(filename, "wb") as file:
            file.write(response.content)
        #print(f"{name} File downloaded successfully.")
    else:
        print(f"{name} Failed to download the file.")
```

```python
import requests

url = "https://cells.ucsc.edu/cortex-dev/exprMatrix.tsv.gz"
filename = "exprMatrix.tsv.gz"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Open a file in binary mode and write the response content
    with open(filename, "wb") as file:
        file.write(response.content)
    print("File downloaded successfully.")
else:
    print("Failed to download the file.")
```

```python
import requests
from bs4 import BeautifulSoup
import os

url = "https://cells.ucsc.edu/cortex-dev/"

# Send a request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all the links on the page
    links = soup.find_all('a')
    
    # Iterate through each link
    for link in links:
        file_url = link.get('href')
        
        # Check if the link is a file
        if file_url.endswith('.txt') or file_url.endswith('.csv'):
            file_response = requests.get(url + file_url)
            
            # Check if the file request was successful
            if file_response.status_code == 200:
                # Specify the path where you want to save the downloaded files
                save_path = "path/to/save/files/"
                
                # Create the directory if it doesn't exist
                os.makedirs(save_path, exist_ok=True)
                
                # Extract the file name from the URL
                file_name = file_url.split('/')[-1]
                
                # Save the file to the specified path
                with open(os.path.join(save_path, file_name), "wb") as file:
                    file.write(file_response.content)
                    
                print(f"File '{file_name}' downloaded successfully.")
            else:
                print(f"Failed to download the file '{file_url}'.")
else:
    print("Failed to access the URL.")
```
